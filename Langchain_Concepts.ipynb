Example Selectors

```python
from langchain.prompts.example_selector import SemanticSimilarityExampleSelector
from langchain.vectorstores import Chroma
from langchain.embeddings import OpenAIEmbeddings
from langchain.prompts import FewShotPromptTemplate, PromptTemplate

example_prompt = PromptTemplate(
    input_variables=["input", "output"],
    template="Example Input: {input}\nExample Output: {output}",
)

examples = [
    {"input" : "chatGPT", "output" : "OpenAI",
     "input" : "AWS", "output" : "Amazon",
     "input" : "Gemini", "output" : "Google",
     "input" : "Iphone", "output" : "Apple",
     "input" : "Whatsapp", "output" : "Meta"}
]
```

```python
!pip install chromadb
```

```python
selector = SemanticSimilarityExampleSelector.from_examples(
    examples,
    OpenAIEmbeddings(openai_api_key=api_key),
    Chroma,
    k=2
)
```

```python
prompt = FewShotPromptTemplate(
    example_selector=selector,
    example_prompt=example_prompt,
    prefix="Tell me to what company it belongs",
    suffix="Input: {noun}\nOutput:",

    input_variables=["noun"],)

my_noun = "VertexAI"

print(prompt.format(noun=my_noun))
```

```python
llm(prompt.format(noun=my_noun))
```

Output Parser

```python
from langchain.output_parsers import StructuredOutputParser, ResponseSchema

response_schemas = [
    ResponseSchema(name="bad", description="This is a badly written and formatted answer"),
    ResponseSchema(name="good", description="This is a good answer")
]

output_parser = StructuredOutputParser.from_response_schemas(response_schemas)
```

```python
format_instructions = output_parser.get_format_instructions()
format_instructions
```

```python
template = """
You will be given a text that is poorly formatted and contains errors in it from the user. Reformat it and fix all errors in it
{format_instructions}
%user input: {user_input}
%Your response:
"""

prompt = PromptTemplate(
    template=template,
    input_variables=["text"],
    partial_variables={"format_instructions": format_instructions}
)

final_prompt = prompt.format(user_input = " I wove rainly veather")
final_prompt
```

```python
res = llm(final_prompt)
res
```

```python
output_parser.parse(res)
```

```python
from langchain.vectorstores import FAISS
from langchain.embeddings import OpenAIEmbeddings
from langchain.document_loaders import TextLoader

loader = TextLoader('worked.txt')
documents = loader.load()

text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=50)

texts = text_splitter.split_documents(documents)

embeddings = OpenAIEmbeddings(openai_api_key=api_key)

db = FAISS.from_documents(texts, embeddings)
```

```python
retriever = db.as_retriever()

docs = retriever.invoke("what does author wanted to study?")

print("\n\n".join([x.page_content[:200] for x in docs[:2]]))
```

Chains

Sequential Chain

```python
from langchain.llms import OpenAI
from langchain.chains import LLMChain
from langchain.prompts import PromptTemplate
from langchain.chains import SimpleSequentialChain

llm = OpenAI(temperature=0.7, openai_api_key=api_key)

template1 = """
You are a cookbook guide that suggest user a food dish based on their location.
% User Location: {location}
% response :
"""

prompt1 = PromptTemplate(
    input_variables=["location"],
    template=template1
)

chain1 = LLMChain(llm=llm, prompt=prompt1)

template2 = """
Given a meal, your job is to provide with a short and simple recipe for that meal.
%meal: {meal}
%response:
"""

prompt2 = PromptTemplate(
    input_variables=["meal"],
    template=template2
)

chain2 = LLMChain(llm=llm, prompt=prompt2)

overall_chain = SimpleSequentialChain(chains=[chain1, chain2], verbose=True)
```

```python
res = overall_chain.run("Nice")
```
